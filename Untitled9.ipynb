```python
#Group by race
df.groupby(['race'])['patient_nbr'].count()
```

```python
#Since there are ? value, I would like to find out how many missing value. I replace "?" to NaN.
df = df.replace("?", np.NaN, )
```

```python
df.isnull().sum()
```

```python
#Group by Gender 
df.groupby(['gender'])['patient_nbr'].count()
```

```python
#Group by Gender and age
df.groupby(['gender','age'])['patient_nbr'].count()
```

```python
#Female readmitted <30 days

df2 = (df['gender'] == 'Female') & (df['readmitted'] == '<30')

df[df2]
```

```python
#drop all unnecessary columns
df4 = df[df2].drop(df.loc[:, 'weight':'diabetesMed'], axis=1)
df4
```

```python
#group by female and age who readmitted <30 days
df5=df4.groupby(['gender','age'])['encounter_id'].count()
df5
```

```python
#graph on which female age group has the most readmitted <30 days rate
df5.plot(kind='bar')
```

```python
#male readmitted <30 days

df6 = (df['gender'] == 'Male') & (df['readmitted'] == '<30')
df[df6]
```

```python
#drop all unnecessary columns
df7 = df[df6].drop(df.loc[:, 'weight':'diabetesMed'], axis=1)
df7
```

```python
#group by male and age who readmitted <30 days
df8=df7.groupby(['gender','age'])['encounter_id'].count()
df8
```

```python
#graph on which male age group has the most readmitted <30 days rate
df8.plot(kind='bar')
```

```python
#The age group[70-80] in both gender has the highest readmitted within a month rate
#One interesting thing I found is Male in the [60-70]age group are more likely to readmit whereas Female in the [80-90]age group are readmit
```
