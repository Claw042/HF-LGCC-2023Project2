```python
#View only Demographic Category: All students in 20th distact
df2= df['School DBN'].str.contains('20K') & (df['Demographic Category'] == 'All Students') & (df['School Level'] == 'Junior High-Intermediate-Middle')
df[df2]
```

```python
# Remove unnecessary columns 
df3 = df[df2].drop(['School Type','Demographic Variable'], axis=1)
df3
```

```python
#Groupby
df3.groupby(['School Name','Year','Regents Exam'])['Mean Score'].max()
```

```python
#Would like to know which school has the highest Mean Algebra Score in 2016

df3['Mean Score'].max()
```

```python
#Since there is 's' in mean score, I would like to remove the S. 

df4 = df3.drop(index=df3[df3['Mean Score']== 's'].index)
df4
```

```python
#Filter to only see Algebra in 2016
df5=df4.loc[(df4['Regents Exam'] == 'Common Core Algebra') & (df4['Year'] == 2016)]
```

```python
df5
```

```python
#Graph on how many student took the Alebra test in 2016
df5.sort_values(by='Mean Score').plot(kind='bar',x='School Name',y= 'Total Tested')
```
