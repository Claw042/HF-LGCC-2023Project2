# Filter columns

```python
#Would like to compare which crime is the highest in Precint 108 and 78
#filter for precint 108 
Precinct_108= df[df['ARREST_PRECINCT'] == 108].sort_values(by='ARREST_DATE')
Precinct_108
```

```python
#filter for precint 78
Precinct_78=df[df['ARREST_PRECINCT'] == 78].sort_values(by='ARREST_DATE')
Precinct_78
```

```python
Precinct_78.groupby(['OFNS_DESC'])['ARREST_PRECINCT'].count()
```

```python
Precinct_78_crime=Precinct_78.groupby(['OFNS_DESC'])['ARREST_PRECINCT'].count().sort_values(ascending=False).plot(kind='bar')

Precinct_78_crime.set_title('Which crime is the highest in Precint 78?')
```

```python
Precinct_108.groupby(['OFNS_DESC','PERP_SEX'])['ARREST_PRECINCT'].count()
```

```python
Precinct_108_crime = Precinct_108.groupby(['OFNS_DESC'])['ARREST_PRECINCT'].count().sort_values(ascending=False).plot(kind='bar')

Precinct_108_crime.set_title('Which crime is the highest in Precint 108?')
```

```python
#wanted to see the crime by gender in Precinct108
sns.jointplot(x='PERP_SEX', y='OFNS_DESC', data= Precinct_108)
```

```python
#wanted to see "ASSAULT 3 & RELATED OFFENSES" by gender in Precinct108

Precinct_108_highest_crime= Precinct_108[Precinct_108['OFNS_DESC']=='ASSAULT 3 & RELATED OFFENSES']

Precinct_108_highest_crime.groupby(['PERP_SEX','AGE_GROUP'])['OFNS_DESC'].count().plot(kind='bar')
```

```python
#wanted to see the crime by gender in Precinct78
sns.jointplot(x='PERP_SEX', y='OFNS_DESC', data= Precinct_78)
```

```python
#wanted to see "ASSAULT 3 & RELATED OFFENSES" by gender in Precinct78

Precinct_78_highest_crime= Precinct_78[Precinct_78['OFNS_DESC']=='PETIT LARCENY']

Precinct_78_highest_crime.groupby(['PERP_SEX','AGE_GROUP'])['OFNS_DESC'].count().plot(kind='bar')
```

# Conclusion

Assault 3 & related offenses are the highest crime in Precint 108. On the other hand, PETIT LARCENY is the highest crime in Precint 78. In both Precint, both highest crime are committed by male at the age group of 25-44.
